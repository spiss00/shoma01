```python
import numpy as np
import matplotlib.pyplot as plt

sigma = 1.5
beta = 0.98
rho = 0.6
sigma_eps = 0.6
r = 0.04
w = 1

hL, hH = np.exp(-sigma_eps), np.exp(sigma_eps)

def u(c):
    return (c**(1-sigma)) / (1-sigma)

def budget_constraint(a_t, h_t):
    return w * h_t + (1 + r) * a_t

def savings_rate(a_t, h_t):
    c_t = budget_constraint(a_t, h_t)
    return (c_t - a_t) / c_t

a_t_values = np.linspace(0, 10, 100)
savings_rates_hL = [savings_rate(a, hL) for a in a_t_values]
savings_rates_hH = [savings_rate(a, hH) for a in a_t_values]

plt.plot(a_t_values, savings_rates_hL, label='hL')
plt.plot(a_t_values, savings_rates_hH, label='hH')
plt.xlabel('Current Assets')
plt.ylabel('Savings Rate')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

sigma = 1.5
beta = 0.98
rho = 0.6
sigma_eps = 0.6
r = 0.04
w = 1
tau_k = 0.3

hL, hH = np.exp(-sigma_eps), np.exp(sigma_eps)

def u(c):
    return (c**(1-sigma)) / (1-sigma)

def budget_constraint_no_tax(a_t, h_t):
    return w * h_t + (1 + r) * a_t

def budget_constraint_with_tax(a_t, h_t):
    return w * h_t + (1 + (1 - tau_k) * r) * a_t

def savings_rate(a_t, h_t, with_tax=False):
    if with_tax:
        c_t = budget_constraint_with_tax(a_t, h_t)
    else:
        c_t = budget_constraint_no_tax(a_t, h_t)
    return (c_t - a_t) / c_t

a_t_values = np.linspace(0, 10, 100)
savings_rates_hL_no_tax = [savings_rate(a, hL) for a in a_t_values]
savings_rates_hH_no_tax = [savings_rate(a, hH) for a in a_t_values]
savings_rates_hL_with_tax = [savings_rate(a, hL, True) for a in a_t_values]
savings_rates_hH_with_tax = [savings_rate(a, hH, True) for a in a_t_values]

plt.plot(a_t_values, savings_rates_hL_no_tax, label='hL without tax')
plt.plot(a_t_values, savings_rates_hH_no_tax, label='hH without tax')
plt.plot(a_t_values, savings_rates_hL_with_tax, label='hL with tax')
plt.plot(a_t_values, savings_rates_hH_with_tax, label='hH with tax')
plt.xlabel('Current Assets)')
plt.ylabel('Savings Rate')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

sigma = 1.5
beta = 0.98
rho = 0.6
sigma_eps = 0.6
r = 0.04
w = 1
T = 1

hL, hH = np.exp(-sigma_eps), np.exp(sigma_eps)

def u(c):
    return (c**(1-sigma)) / (1-sigma)

def budget_constraint_no_subsidy(a_t, h_t):
    return w * h_t + (1 + r) * a_t

def budget_constraint_with_subsidy(a_t, h_t):
    return w * h_t + (1 + r) * a_t + T

def savings_rate(a_t, h_t, with_subsidy=False):
    if with_subsidy:
        c_t = budget_constraint_with_subsidy(a_t, h_t)
    else:
        c_t = budget_constraint_no_subsidy(a_t, h_t)
    return (c_t - a_t) / c_t

a_t_values = np.linspace(0, 10, 100)
savings_rates_hL_no_subsidy = [savings_rate(a, hL) for a in a_t_values]
savings_rates_hH_no_subsidy = [savings_rate(a, hH) for a in a_t_values]
savings_rates_hL_with_subsidy = [savings_rate(a, hL, True) for a in a_t_values]
savings_rates_hH_with_subsidy = [savings_rate(a, hH, True) for a in a_t_values]

plt.plot(a_t_values, savings_rates_hL_no_subsidy, label='hL without subsidy')
plt.plot(a_t_values, savings_rates_hH_no_subsidy, label='hH without subsidy')
plt.plot(a_t_values, savings_rates_hL_with_subsidy, label='hL with subsidy')
plt.plot(a_t_values, savings_rates_hH_with_subsidy, label='hH with subsidy')
plt.xlabel('Current Assets')
plt.ylabel('Savings Rate')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

sigma = 1.5
beta_original = 0.98
beta_new = 0.1
rho = 0.6
sigma_eps = 0.6
r = 0.04
w = 1

hL, hH = np.exp(-sigma_eps), np.exp(sigma_eps)

def u(c, beta):
    return (c**(1-sigma)) / (1-sigma)

def budget_constraint(a_t, h_t, beta):
    return w * h_t + (1 + r) * a_t

def savings_rate(a_t, h_t, beta):
    c_t = budget_constraint(a_t, h_t, beta)
    return (c_t - a_t) / c_t

a_t_values = np.linspace(0, 10, 100)
savings_rates_hL_original = [savings_rate(a, hL, beta_original) for a in a_t_values]
savings_rates_hH_original = [savings_rate(a, hH, beta_original) for a in a_t_values]
savings_rates_hL_new = [savings_rate(a, hL, beta_new) for a in a_t_values]
savings_rates_hH_new = [savings_rate(a, hH, beta_new) for a in a_t_values]

plt.plot(a_t_values, savings_rates_hL_original, label='hL (beta=0.98)')
plt.plot(a_t_values, savings_rates_hH_original, label='hH (beta=0.98)')
plt.plot(a_t_values, savings_rates_hL_new, label='hL (beta=0.1)')
plt.plot(a_t_values, savings_rates_hH_new, label='hH (beta=0.1)')
plt.xlabel('Current Assets (a_t)')
plt.ylabel('Savings Rate')
plt.legend()
plt.show()
```
